**모델링 준비**

```python
# 지역, 가격(중품) 열을 제외한다.
frame1 = frame1.drop(['location', 'priceMid'], axis=1)
frame2 = frame2.drop(['location', 'priceMid'], axis=1)
frame3 = frame3.drop(['location', 'priceMid'], axis=1)
frame4 = frame4.drop(['location', 'priceMid'], axis=1)
frame5 = frame5.drop(['location', 'priceMid'], axis=1)

# 토마토의 경우 모든 지역에서 생산량에 결측값이 존재하므로 해당 열을 제외한다.
# frame1 = frame1.drop('production', axis=1)
# frame2 = frame2.drop('production', axis=1)
# frame3 = frame3.drop('production', axis=1)
# frame4 = frame4.drop('production', axis=1)
# frame5 = frame5.drop('production', axis=1)
```

```python
# 날짜열의 형식을 변환한다.
frame1['date'] = pd.to_datetime(frame1['date'], infer_datetime_format=True)
frame2['date'] = pd.to_datetime(frame2['date'], infer_datetime_format=True)
frame3['date'] = pd.to_datetime(frame3['date'], infer_datetime_format=True)
frame4['date'] = pd.to_datetime(frame4['date'], infer_datetime_format=True)
frame5['date'] = pd.to_datetime(frame5['date'], infer_datetime_format=True)

from datetime import datetime
frame1['date'] = frame1['date'].apply(datetime.toordinal)
frame2['date'] = frame2['date'].apply(datetime.toordinal)
frame3['date'] = frame3['date'].apply(datetime.toordinal)
frame4['date'] = frame4['date'].apply(datetime.toordinal)
frame5['date'] = frame5['date'].apply(datetime.toordinal)
```

```python
# 모델 학습을 위해 데이터를 train/test set으로 분리한다.
from sklearn.model_selection import train_test_split

# 원하는 지역 선택
df = frame1.copy()

X = df.iloc[:,2:].values
y = df.priceHigh.values
feature_names = df.iloc[:,2:].columns
X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
print(feature_names)
```

## 2 Random forest regression with sklearn

**모델링 및 학습**

```python
from sklearn.ensemble import RandomForestRegressor

# 모델 초기화 및 train set으로 학습
rfr = RandomForestRegressor(n_estimators=1000, random_state=1234)
rfr.fit(X_train, y_train)
```

```python
# R-squared 값 출력
print("The R-squared value of train set is: {0:0.4f} \n".format(rfr.score(X_train, y_train)))
print("The R-squared value of test set is: {0:0.4f} \n".format(rfr.score(X_test, y_test)))
```

**예측**

```python
# 가격 예측
# 다른 지역의 데이터로도 테스트해본다.
newX = X
newY = y

predicted = rfr.predict(newX)
predictedDf = pd.DataFrame({"predicted": predicted})
# dataframe for X
newXDf = pd.DataFrame(newX)
newXDf.columns = feature_names
# join X, y and prediction
mydata_with_pd = newXDf.join(predictedDf)
mydata_with_pd.head()
```

```python
# 예측 결과 평가 (MAE, RMSE, RAE, RSE)
import numpy as np
obs = frame5.priceHigh
pred = mydata_with_pd.predicted

mae = np.mean(abs(pred-obs))
rmse = np.sqrt(np.mean((pred-obs)**2))
rae = np.mean(abs(pred-obs))/np.mean(abs(obs-np.mean(obs)))
rse = np.mean((pred-obs)**2)/np.mean((obs-np.mean(obs))**2)

print("Mean Absolute Error: {0:0.6f}".format(mae))
print("Root Mean Squared Error: {0:0.6f}".format(rmse))
print("Relative Absolute Error: {0:0.6f}".format(rae))
print("Relative Squared Error: {0:0.6f}".format(rse))
```

```python
d = {"Mean Absolute Error": mae, "Root Mean Squared Error": rmse, "Relative Absolute Error": rae, "Relative Squared Error": rse}
evl = pd.DataFrame([d])
```
